```python
import pandas as pd
import numpy as np
import pmdarima
import arch
import warnings
warnings.filterwarnings("ignore")
import matplotlib.pyplot as plt 
plt.style.use('tableau-colorblind10')
plt.rcParams['figure.figsize'] = (16, 5)
plt.rcParams['figure.dpi'] = 300

df = pd.read_csv('./nasdaq_d_2012_2022.csv')
```

```python
dt = df.pivot(index="date",columns="tic", values="adjcp").fillna(method="bfill") # transform to the standard format
```

```python
dt.plot(legend = False)
plt.savefig('./allstock.jpg')
```

```python
dt.pct_change().dropna().plot(legend = False)
plt.savefig('./return.jpg')
```
